```python
import pandas as pd
import numpy as np
```

```python
school={'Name':['Sam','Jacob',
                'Smith','Ben',
                'Jaden','Keir',
                'Barron','Jeff',
                'James','Michael','Kingston'],
       'Roll_No.':[9,5,
                   3,2,
                   5,1,
                   7,8,
                   3,6,4],
       'Homework':pd.Series(['Done',np.nan,
                             'Not Done','Done',
                            np.nan,'Not Done',
                            'Not Done',np.nan,
                            'Done',np.nan,'Not Done']),
       'Today_Attendance':pd.Series(['Absent','Present',
                                     'Present','Present',
                                    'Absent','Absent',
                                     'Present','Absent',
                                     'Present','Absent','Absent']),
       'Days_of_Absent':[2,1,np.nan,4,8,1,9,5,7,4,6],
       'Marks':[50,40,65,23,19,84,75,9,13,34,5]}
df=pd.DataFrame(school)
df
```

# Sorting

## Sorting is the process of arranging items or data in a specific order, such as alphabetically, numerically, and ordering data in sequence like ascending and descending.

```python
df[['Name','Marks']].sort_values(by='Marks')
```

```python
df[['Name','Marks']].sort_values(by='Marks',ascending=True)[0:4]
```

```python
df[['Name','Marks']].sort_values(by='Marks',ascending=False)[0:4]
```
